```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

print("Hello World")
```

```python
CLUB_LIST = glob.glob("*.csv"); CLUB_LIST
```

```python
clubDf = {}

for i in CLUB_LIST:
    df = pd.read_csv(i, encoding='cp949')
#     print(a)
#     df = df.drop(["순번", "지원 직종", "학과", "면접시간", "연락처", "비고"], axis=1)
    df = df[["학번", "이름"]]
    df = df.dropna()
    clubDf[i.split(".")[0]] = df
    
clubDf
```

```python
clubTotal = pd.DataFrame(columns = ['ID', 'name', 'club'])
# clubTotal
for key, value in list(clubDf.items()):
    print(f"{key}, 총 {len(value)}명")
    clubCount = 0
    for i in value.index:
        student = {"ID" : value["학번"][i], "name" : value["이름"][i], "club" : key}
        clubTotal = clubTotal.append(student, ignore_index=True)

clubTotal    
```

```python
swStudentCondition = (10400 < clubTotal["ID"]) & (clubTotal["ID"] < 10699)
cdStudentCondition = (11000 < clubTotal["ID"]) & (clubTotal["ID"] < 11299)

swStudentDf = clubTotal[swStudentCondition]
cdStudentDf = clubTotal[cdStudentCondition]
```
