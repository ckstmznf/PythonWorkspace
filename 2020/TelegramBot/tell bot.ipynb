```python
import requests as re
import bs4 as bs
import pandas as pd

def melon(count):
    url = "https://www.melon.com/chart/index.htm"
    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
    req = re.get(url, headers = header).text
    soup = bs.BeautifulSoup(req)
    datas = soup.find("div", class_="service_list_song type02 no_rank d_song_list").find_all("tr")[1:]
    text = ""
    find_datas = []
    for n, data in enumerate(datas):
        real_data = data.find_all("td")
        title = real_data[3].find("div", class_="ellipsis rank01").find("a").text
        artst = real_data[3].find("div", class_="ellipsis rank02").find("a").text
    #     print(f"top{n+1} {title}, {artst}")
#         find_datas.append({"title" : title, "artst" : artst})
        text += f"top {n + 1} : {title}, {artst}\n"
        if n + 1 == count :
            break
    return text
```

# 네이버 웹툰 검색

```python
# 네이버 웹툰 검색
def naver_webtoon_search_api(key):
    url = "https://comic.naver.com/search.nhn?keyword=" + key
    soup = make_soup(url)

    data = soup.select(".resultBox")[0].select(".resultList>li")
    res = []
    bool_ = False
    for n, i in enumerate(data):
    #     if n == 1:
    #         print(i)
    #         break
        title = i.find("h5").find("a").text
        for j in res:
            if j["title"] == title:
                bool_ = True
                break
                
        if not bool_:
            link = i.find("h5").find("a")["href"].split("?")[1].split("&")[0].split("=")[1]
            data_2 = i.find_all("li")[0]
            writer = data_2.find("a").text.strip()
            res.append({"title" : title, "writer" : writer, "link" : link})
        else:
            bool_ = False
        
        
    if not res:
        return "검색 결과가 없습니다."
    return res


def naver_webtoon_search(key):
    key = "마음"
    text = f"{key}에 대한 검색 결과 입니다.\n"
    data = naver_webtoon_search_api(key)
    for i in data:
        text += f"▶ 제목 : {i['title']}, 작가 : {i['writer']}\n링크 : https://comic.naver.com/webtoon/list.nhn?titleId={i['link']}\n\n"
    
    return text
```
